```python
sns.countplot(data=df, x="class");
```

```python
feat_uni = df.describe().T.reset_index().sort_values("unique")
```

```python
plt.figure(figsize=(12,8),dpi=100)
sns.barplot(data=feat_uni, x="index", y="unique")
plt.xticks(rotation=90);
```

```python
X = df.drop("class", axis=1)
```

```python
X.isnull().sum()
```

```python
X = pd.get_dummies(X, drop_first=True)
```

```python
X
```

```python
y = df["class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=101)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {"n_estimators":[50,100], "learning_rate":[0.05,0.1,0.2], "max_depth":[3,4,5]}
```

```python
gb_model = GradientBoostingClassifier()
```

```python
grid = GridSearchCV(gb_model,param_grid)
```

```python
grid.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report, plot_confusion_matrix, accuracy_score
```

```python
predictions = grid.predict(X_test)
```

```python
grid.best_estimator_
```

```python
grid.best_params_
```

```python
print(classification_report(y_test,predictions))
```

```python
plot_confusion_matrix(grid,X_test,y_test);
```

```python
grid.best_estimator_.feature_importances_
```

```python
feat_import = grid.best_estimator_.feature_importances_
```

```python
imp_feat = pd.DataFrame(index=X.columns, data=feat_import, columns=["Importance"])
```

```python
imp_feat = imp_feat[imp_feat["Importance"]>0.0005].reset_index()
imp_feat.sort_values("Importance")
```

```python
plt.figure(figsize=(12,8),dpi=100)
sns.barplot(data=imp_feat.sort_values("Importance"),x="index",y="Importance")
plt.xticks(rotation=90);
```
