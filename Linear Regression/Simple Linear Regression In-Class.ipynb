##  Simple Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Advertising.csv")
df.head()
```

```python
df["total_spend"] = df.TV + df.radio + df.newspaper
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="total_spend", y="sales", data=df);
```

```python
df.sales.corr(df.total_spend)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(x=df.total_spend, y=df.sales, ci=None);
```

```python
# y = b1x + b0
```

```python
X = df["total_spend"]
y = df["sales"]
```

```python
b1, b0 = np.polyfit(X, y, deg=1)
print(b1,b0)
```

```python
potential_spend = np.linspace(0, 500, 100) #theoretical potential spend
```

```python
potential_spend
```

```python
predicted_sales = b1 * potential_spend + b0
```

```python
predicted_sales
```

```python
plt.figure(figsize=(10,6))
plt.plot(potential_spend,predicted_sales, color="red")
sns.scatterplot(x=df.total_spend, y=df.sales);
```

```python
# y = B3x**3 + B2x**2 + B1x + B0
```

```python
b3, b2, b1, b0 = np.polyfit(X,y,deg=3)
print(b3, b2, b1, b0)
```

```python
pot_spend = np.linspace(0,500,100)
```

```python
pred_sales = b3 * pot_spend ** 3 + b2 * pot_spend ** 2 + b1 * pot_spend + b0
```

```python
plt.figure(figsize=(10,6))
plt.plot(pot_spend,pred_sales, color="red")
sns.scatterplot(x=df.total_spend, y=df.sales);
```
